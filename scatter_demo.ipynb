Scatter demo

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from freeplot.base import FreePlot
from scipy.stats import multivariate_normal
```

```python
%matplotlib inline
```

```python
titles = ("S", "h", "a", "n")
labels = ("sin", "cos", "x")
```

```python
fp = FreePlot((1, 4), (9.5, 2), titles=titles, dpi=100, sharey=True)
```

```python
normal = multivariate_normal

nums = 100
means = (
    (0, 0),
    (5, 5),
    (-5, -5)
)

cov = (
    2,
    1,
    1
)
```

```python
# plot for each
for title in titles:
    for i, mean in enumerate(means):
        data = normal.rvs(mean, cov[i], size=nums)
        fp.scatterplot(data[:, 0], data[:, 1], index=title, label=labels[i])
```

```python
fp.set_title(y=1.) # set title for each
fp.set_label("Y", index=0, axis='y') # set y label as "Y" for the first plot
fp.set_label("X", index=None, axis='x') # set x label as "X" for all plots
fp[0].legend() # set legend on the first plot
```
