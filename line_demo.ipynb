```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from freeplot.base import FreePlot
```

```python
%matplotlib inline
```

```python
titles = ("S", "h", "a", "n")
labels = ("sin", "cos", "x")
```

```python
fp = FreePlot((1, 4), (9.5, 2), titles=titles, dpi=100, sharey=True) # shape: (1, 4), figsize: (9.5, 2)
```

```python
# create data
nums = 20
x = np.linspace(-10, 10, nums)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x
ys = (y1, y2, y3)
```

```python
# plot for each
for title in titles:
    for i, y in enumerate(ys):
        y = y + np.random.randn(nums)
        fp.lineplot(x, y, index=title, label=labels[i])
```

```python
fp.set_title(y=1.) # set title for each
fp.set_label("Y", axis='y') # set y label for the first plot
fp.set_label("X", index=None, axis='x') # set x label for all plots
fp[0].legend() # add legend at the first plot
```
