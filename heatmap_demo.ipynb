```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from freeplot.base import FreePlot
```

```python
%matplotlib inline
```

```python
titles = ("S", "h", "a", "n")
row_labels = ('c', 'u', 't', 'e')
col_labels = ('l', 'r', 'i', 'g')
```

```python
fp = FreePlot((1, 4), (9, 2), titles=titles, dpi=100, sharey=True)
```

```python
for title in titles:
    data = np.random.rand(4, 4)
    df = pd.DataFrame(data, index=col_labels, columns=row_labels)
    fp.heatmap(df, index=title, annot=True, fmt=".4f", cbar=False, linewidth=0.5)
```

```python
fp.set(xlabel="X") # set x label as "X" for all plots
fp.set_label('Y', index=0, axis='y')
```
